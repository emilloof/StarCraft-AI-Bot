```python
import json

def load_and_analyze_json(file_path):
  # Load the JSON file
  with open(file_path) as file:
    data = json.load(file)

  # Collect statistics
  stats = {
    'total_enemies': len(data),
    'mean_value': sum(item[1] for item in data) / len(data)
  }

  # Print the statistics
  print(stats)

print("other:")
load_and_analyze_json('stats\\together\\normal_scout_enemies_3min_computer_terran.json')
load_and_analyze_json('stats\\together\\pfscout_enemies_3min_computer_terran.json')


print("against self:")
load_and_analyze_json('normal_scout_enemies_3_10min_basicagent_terran.json')
load_and_analyze_json('pfscout_enemies_3_30min_basicagent_terran.json')

print("new:")
load_and_analyze_json('pfscout_easy_protoss_2-45_enemies.json')
load_and_analyze_json('normal_scout_easy_protoss_2-45_enemies.json')
```
